```python
from nltk import pos_tag, word_tokenize
def posTag(review):
    tagged_text_list=[]
    for text in review:
        tagged_text_list.append(pos_tag(word_tokenize(text)))
    return tagged_text_list
#posTag("this is random text")
df1['tagged_text']=posTag(df1.clean_text)
df2['tagged_text']=posTag(df2.clean_text)
df3['tagged_text']=posTag(df3.clean_text)
df4['tagged_text']=posTag(df4.clean_text)
df5['tagged_text']=posTag(df5.clean_text)
df6['tagged_text']=posTag(df6.clean_text)
df7['tagged_text']=posTag(df7.clean_text)
df8['tagged_text']=posTag(df8.clean_text)
```

```python
def filterTag(tagged_review):
    final_text_list=[]
    for text_list in tagged_review:
        final_text=[]
        for word,tag in text_list:
            if tag in ['NN','NNS','NNP','NNPS','RB','RBR','RBS','JJ','JJR','JJS','VB','VBD','VBG','VBN','VBP','VBZ']:
                final_text.append(word)
        final_text_list.append(' '.join(final_text))
    return final_text_list

df1['final_text']=filterTag(df1.tagged_text)
df2['final_text']=filterTag(df2.tagged_text)
df3['final_text']=filterTag(df3.tagged_text)
df4['final_text']=filterTag(df4.tagged_text)
df5['final_text']=filterTag(df5.tagged_text)
df6['final_text']=filterTag(df6.tagged_text)
df7['final_text']=filterTag(df7.tagged_text)
df8['final_text']=filterTag(df8.tagged_text)
```

```python
w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
lemmatizer = nltk.stem.WordNetLemmatizer()

def lemmatize_text(text):
    return [lemmatizer.lemmatize(w) for w in w_tokenizer.tokenize(text)]

df1['text_lemmatized'] = df1.final_text.apply(lemmatize_text)
df2['text_lemmatized'] = df2.final_text.apply(lemmatize_text)
df3['text_lemmatized'] = df3.final_text.apply(lemmatize_text)
df4['text_lemmatized'] = df4.final_text.apply(lemmatize_text)
df5['text_lemmatized'] = df5.final_text.apply(lemmatize_text)
df6['text_lemmatized'] = df6.final_text.apply(lemmatize_text)
df7['text_lemmatized'] = df7.final_text.apply(lemmatize_text)
df8['text_lemmatized'] = df8.final_text.apply(lemmatize_text)
```
